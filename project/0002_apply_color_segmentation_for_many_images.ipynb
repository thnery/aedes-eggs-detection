# Applying the color segmentation to other samples

This script applies the color segmentation 0001_color_segmentation code in another image samples.
The results were not so good at all.

```python
import os

import numpy as np
import cv2 as cv

from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from matplotlib import colors

plt.rcParams['figure.dpi'] = 150

print("OpenCV Version: %s" % cv.__version__)
```

```python
light = (110, 0, 0)
dark = (255, 185, 120)
directory = './images/eggs/'
for filename in os.listdir(directory):
    image = cv.imread(directory + filename)
    image = cv.cvtColor(image, cv.COLOR_BGR2RGB)
    hsv_image = cv.cvtColor(image, cv.COLOR_RGB2HSV)
    
    mask = cv.inRange(hsv_image, light, dark)
    result = cv.bitwise_and(image, image, mask=mask)

    plt.subplot(1, 2, 1)
    plt.imshow(mask)
    plt.subplot(1, 2, 2)
    plt.imshow(result)
    plt.show()    
```
